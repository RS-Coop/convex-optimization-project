```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
data = pd.read_csv('./spambase/spambase_train.data', header=None)

data_map = lambda x: np.log(x+0.1)
```

```python
data.head()
```

```python
X = data.iloc[:, 0:-1].apply(data_map)
y = data.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
ds = tf.data.Dataset.from_tensor_slices((X, y))
```
